```python
import matplotlib.pyplot as plt
```

```python
pt=[60,65,68,60,66,62,63]
pt2=[61,62,66,65,69,61,65]
pt3=[60,63,65,68,62,63]
```

```python
plt.plot(pt)
plt.plot(pt2)
plt.plot(pt3)
plt.show()
```

```python
plt.plot(pt, c='r', marker='o', mec='g', mfc='y')
plt.plot(pt2, c='b', marker='X', mec='g', mfc='y')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.title('sample vs value')
plt.legend(['DBMS','CN'])
plt.show()
```

```python
plt.plot(pt, c='r', marker='o', ls='--', mec='g', mfc='y')
plt.plot(pt2, c='b', marker='X', ls=':', mec='k', mfc='C6')
plt.plot(pt3, c='k', marker='s', ls="-.", mec='c', mfc='m')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.title('Student Exam Performance')
plt.legend(['DBMS','CN','DSA'])
plt.show()
```
